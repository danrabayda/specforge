```python
from specforge import CustomFeatureExtractor, load_waveform, window_generator

sample_rate = 4000

extractor = CustomFeatureExtractor.from_sample_rate(
    sample_rate=sample_rate,
    window_length=5,
    spec_freq_dim=256,
    spec_time_dim=96,
)

waveform = load_waveform("../assets/test.wav", sample_rate)

windows = window_generator(
    waveform,
    overlap=0.5,
    window_length=5,
    sample_rate=sample_rate,
)

features = extractor(windows)

print(features["input_values"].shape)
```
